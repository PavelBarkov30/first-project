```python
# Load packages
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as sts
import random
import seaborn as sns
```

# Непрерывное распределение

Сгенерируем выборку объёма 1000 из непрерывного логнормального гамма-распределения.

```python
disttype = sts.loggamma(1.)
sample = disttype.rvs(1000)
```

Напишем функцию для тестирования выборки на нормальность. Для проверки выборки удовлетвряет ли она нормальному распределению будем использовать стат. тесты Колмогорова-Смирнова и Шапиро-Уилкоксона:

```python
def check_norm_distribution(vect):
    # Запускаем тест Колмогорова-Смирнова
    ksNormalcy = sts.kstest(vect, "norm")
    print("Kolmogorov-Smirnov: p-value = " + str(round(ksNormalcy[1], 5)))
    # Запускаем тест Шапиро-Уилкоксона
    shapNormalcy = sts.shapiro(vect)
    print("Shapiro-Wilk: p-value = " + str(round(shapNormalcy[1], 5)))
    # Если p-value тестов меньше 0.05, то гипотеза о нормальности
    # распределения отвергается
    if shapNormalcy[1] < 0.05 and ksNormalcy[1] < 0.05:
         x = False
    else:
        x = True
    return x
```

Построим гистрограмму распределений и функцию плотности на основе KDE,
и выведем среднее и дисперсию.

```python
x = np.linspace(np.min(sample), np.max(sample),100)
pdf = disttype.pdf(x)
plt.plot(x, pdf)
plt.hist(sample, bins=int(len(sample) / 25),
         facecolor='g', alpha=0.75,
         density = True)

tx = "$\mu=" + str(round(np.mean(sample), 2)) + ",\ \sigma=" + str(round(np.std(sample), 2)) + "$"

plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Log Gamma Distribution')
plt.text(-5, 0.3, tx)
plt.grid(True)
plt.show()
```

Проверяем на нормальность распределение. Результат FALSE. 
Распределение ненормальное как и должно быть.

```python
print(check_norm_distribution(sample))
```

Построим распределение на основе выборочных средних с выборкой в размере 5 и снова проверим на нормальность. Результат FALSE. Распределение ненормальное. Но график сдвинулся, распределение перестало быть логнормальным.

```python
# Исследуем влияние количества выборок для получения среднего, проитерировавшись по N.
for n in (5, 10, 50, 100, 1000):
    # Генерим выборку из выборочных средних объема n 
    # из логгамма распределения
    dist = [np.mean(disttype.rvs(n)) for i in range(1000)]
    #sampleList[n] = [np.mean(random.sample(sample,  n)) for i in range(1000)]
    # Считаем дисперсию
    vr = np.var(dist)
    mn = np.mean(dist)
    # Поверх гистограммы нарисуем плотность 
    # соответствующего нормального распределения
    # Посчитаем стандартное отклонение
    sd = np.sqrt(vr)
    # Генерируем нормальное распределение для плотности
    normalDist = sts.norm(mn, sd)
    normalPdf = normalDist.pdf(x)
    
    plt.plot(x, normalPdf)
    plt.hist(dist, bins = int(len(dist) / 25),
             facecolor='g', alpha=0.75, density = True)
    # Добаввим текст с информацией по средней и ср. откл
    # tx = "$\mu=" + str(round(mn, 2)) + ",\ \sigma=" + str(round(sd, 2)) + "$"
    
    plt.xlabel('Value')
    plt.ylabel('Probability')
    plt.title('Log Gamma Dist and Normal Density')
    # plt.text(2, 0.3, tx)
    plt.grid(True)
    plt.show()
    # Будем проверять на нормальность на каждой итерации
    print(check_norm_distribution(dist))
```

Вывод: На примере логнормального гамма распределения мы увидели как работает ЦПТ и спустя n итераций распределение на основе выборочного среднего постепенно стало нормальным. 

Как видно на графиках выше распределение менялось поэтапно и стало нормальным только в конце.
